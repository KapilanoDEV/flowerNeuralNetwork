```python
def cost(b):
    return (b-4) ** 2
```

```python
cost(8)
```

```python
def num_slope(b):
    h = 0.0001
    return (cost(b+h) - cost(b))/h
```

```python
num_slope(5)
```

```python
def slope(b):
    return 2 * (b - 4)
```

```python
slope(3)
```

```python
b = -20
```

```python
b = 0
for i in range(10):
    b = b - .1 * slope(b)
    print(b)
```
